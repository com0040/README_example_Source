```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
url = requests.get('https://finance.naver.com/sise/sise_market_sum.naver?&page=1')
html =  BeautifulSoup(url.text)

table = html.find('table',class_='type_2')
table = pd.read_html(str(table))[0] # -> 리스트안에 데이터에서 [0]번째 데이터를 뽑아온다 - 데이터프레임으로
# -> 누락값이 보인다(실선을 읽어와서)
table = table[table['종목명'].notnull()] # 종목명이 누락값이 아닌 것을 가져온다다

del table['N']
del table['토론실'] # 필요없는 열 삭제

table
```

```python
url = requests.get('https://finance.naver.com/sise/sise_market_sum.naver?&page=1')
html =  BeautifulSoup(url.text)
kospi_page = int(html.find('td',class_='pgRR').find('a')['href'].split('=')[-1])

total =[]

for n in range(1,kospi_page+1):
    # n페이지를 가져오기
    url = requests.get('https://finance.naver.com/sise/sise_market_sum.naver?&page={}'.format(n)) # 요청
    html =  BeautifulSoup(url.text) # 번역 

    table = html.find('table',class_='type_2') # 테이블 찾고 
    table = pd.read_html(str(table))[0] # -> 리스트안에 데이터에서 [0]번째 데이터를 뽑아온다 - 데이터프레임으로
    # -> 누락값이 보인다(실선을 읽어와서)
    table = table[table['종목명'].notnull()] # 종목명이 누락값이 아닌 것을 가져온다

    del table['N']
    del table['토론실'] # 필요없는 열 삭제

    total.append(table)
    time.sleep(1)

    print('{} 페이지 저장완료...'.format(len(total)))
```

```python
kospi =pd.concat(total, ignore_index=True)
kospi
```
