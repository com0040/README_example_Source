```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
url = requests.get('https://search.daum.net/search?w=tot&DA=YZR&t__nil_searchbox=btn&sug=&sugo=&sq=&o=&q=%EB%A1%9C%EB%98%90')
# 200 : 문제없음 / 400번대 : 존재하지 않는 페이지 / 500번대 : 로그인처리. 잠금 /
url
```

```python
url.text
```

```python
html = BeautifulSoup(url.text)
html
```

```python
html.find('span',class_='f_red') # 테그포함 1036회 <-text
```

```python
html.find('span',class_='f_red').text # 테그미포함, text 추출출
```

```python
current=html.find('span',class_='f_red').text.replace('회','') # 테그미포함, text추출, '회' 없에기 = 문자열
```

```python
current=int(html.find('span',class_='f_red').text.replace('회','')) # 테그미포함, text추출, '회' 없에기
current # 정수
```

```python
html.find("div", class_='lottonum') # 로또번호 테그, 클래스=html 반환환
```

```python
lotto = html.find("div", class_='lottonum').find_all('span') # 로또번호 테그, 클래스
# find_all = 리스트 반환
del lotto[-2] # 보너스 삭제
del lotto[-2] # 보너스 삭제

lotto
```

```python
for i in lotto:
    print(i.text) # 로또에서 텍스트만 따서 프린트트
```

```python
box = []
for i in lotto:
    box.append(i.text) # 로또에서 텍스트만 따서 box에 추가 - 문자열
box # 문자열
```

```python
box = []
for i in lotto:
    box.append(int(i.text)) # 로또에서 텍스트만 따서 box에 추가 - 정수 
box # 정수 
```
